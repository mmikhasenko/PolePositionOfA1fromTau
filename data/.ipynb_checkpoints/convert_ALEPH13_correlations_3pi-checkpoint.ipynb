## 2013 ALEPH $\tau \to \nu_\tau \,3\pi$

```julia
f = open("aleph13_3pi.f", "r")
lines = readlines(f)
```

```julia
function convert_block_to_vector(block)
    ll1 = [replace(line,",","") for line in block]
    ll2 = [replace(line,"&","") for line in ll1]
    ll3 = [replace(line,"/","") for line in ll2]
    ll4 = [split(line) for line in ll3]
    vcat([[parse(Float64,n) for n in line] for line in ll4]...)
end
```

```julia
block_s = lines[45:60]
```

```julia
block_binw = lines[62:77]
```

```julia
block_data = lines[79:94]
```

```julia
block_stat = lines[96:111]
```

```julia
block_syst = lines[113:128]
```

```julia
val_s = convert_block_to_vector(block_s)[6:end];
val_data = convert_block_to_vector(block_data)[6:end];
val_binw = convert_block_to_vector(block_binw)[6:end];
val_stat = convert_block_to_vector(block_stat)[6:end];
val_syst = convert_block_to_vector(block_syst)[6:end];
```

```julia
using Plots
```

```julia
val_s[7:end-2]
```

```julia
scatter(val_s)
plot!(10*val_binw)
```

```julia
plot(val_data, yerr=val_stat,lw=0,lab="STAT")
```

```julia
pls = plot(layout=grid(1,2), size = (1000,400))
plot!(pls[1], val_s, val_data./val_binw, yerr=val_stat./val_binw,lw=0,lab="STAT")
plot!(pls[2], val_s, val_data./val_binw, yerr=val_syst./val_binw,lw=0,lab="SYST")
```

```julia
plot(val_s,[val_stat val_syst], lab = ["stat" "syst"])
```

The statistical errors are likely calculated as $\sqrt{N}$ where $N$ is the number of the events in the bin.
Comparing the scaled statinstical error $f\sigma_s$ with the Poisson error $\sqrt{N} = \sqrt{f\,d}$ I find the rate,
which was notmilized out.

```julia
begin
    factor = 2300;
    pls = plot(layout=grid(1,2), size = (1000,400))
    plot!(pls[1], val_s,[factor*val_stat sqrt.(factor*val_data)], lab = ["f*stat" "sqrt(f*data)"])
    bar!(pls[2], val_s,factor*val_data, title = "Original spectrum",c=:green, lab="")
end
```

### Export to a text file

```julia
writedlm("ALEPH.2013.3pi.formatted", hcat(val_s, val_data./val_binw, val_stat./val_binw, val_syst./val_binw))
```

### Statistical error correlation matrix

```julia
Ndp = 80
lstart = 130
lshift = 17
bstat = Vector(Ndp);
for i=1:Ndp
    println(i,": ",lines[(lstart-1)+(i-1)*lshift])
    bstat[i] = lines[((lstart+1)+(i-1)*lshift):((lstart-2)+i*lshift)]
end
```

```julia
mstat = hcat([convert_block_to_vector(bstat[i]) for i in 1:Ndp]...)
```

Statistic correlation matrix

```julia
heatmap(mstat[:,6:end],size=(450,350)) # [v≈100.0 ? 0.0 : v for v in mstat])
```

```julia
savefig("ALEPH2013.stat.pdf")
```
