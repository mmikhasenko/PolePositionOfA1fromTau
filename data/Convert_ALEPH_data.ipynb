## 2005 ALEPH $\tau \to \nu_\tau \,\pi^-2\pi^0$

```julia
f = open("/localhome/mikhasenko/cernbox/current_projects/unitarity/isospin/precalc/ALEPH.2005.pi2pi0", "r")
lines = readlines(f)
```

```julia
ll1 = [replace(line,",","") for line in lines]
ll2 = [replace(line,"&","") for line in ll1]
ll3 = [replace(line,"/","") for line in ll2]
ll4 = [split(line) for line in ll3]
ll5 = [ll4[(29*i+2):(29*(i+1)),:] for i=0:3]
data_100 = hcat([vcat([[parse(Float64,n) for n in line] for line in ll]...) for ll in ll5]...)
```

```julia
plot(data_100[:,1],data_100[:,2])
```

```julia
writedlm("/localhome/mikhasenko/cernbox/current_projects/unitarity/isospin/precalc/ALEPH.2005.pi2pi0.formatted", data_100)
```
