```julia
f = open("/localhome/mikhasenko/Downloads/aleph_3pi.f", "r")
lines = readlines(f)
```

```julia
function convert_block_to_vector(block)
    ll1 = [replace(line,",","") for line in block]
    ll2 = [replace(line,"&","") for line in ll1]
    ll3 = [replace(line,"/","") for line in ll2]
    ll4 = [split(line) for line in ll3]
    vcat([[parse(Float64,n) for n in line] for line in ll4]...)
end
```

```julia
block_s = lines[38:65]
```

```julia
block_data = lines[67:94]
```

```julia
block_stat = lines[96:123]
```

```julia
block_syst = lines[125:152]
```

```julia
val_s = convert_block_to_vector(block_s);
val_data = convert_block_to_vector(block_data);
val_stat = convert_block_to_vector(block_stat);
val_syst = convert_block_to_vector(block_syst);
```

### Get rid of zero values

```julia
neq_cut = [(v!=0) for v in val_data];
Ncd = length(neq_cut[neq_cut])
```

```julia
cval_stat = val_stat[neq_cut]
cval_syst = val_syst[neq_cut]
cval_data = val_data[neq_cut]
cval_s = val_s[neq_cut];
# correlation matrices
cmstat = mstat[neq_cut,neq_cut]
cmsyst = msyst[neq_cut,neq_cut];
```

Find eigenvalues rotation matrix for the statistical covariance

```julia
# statistic
mcov_stat = one(cmstat);
for i=1:size(cmstat,1), j=1:size(cmstat,2)
     mcov_stat[i,j] = cmstat[i,j]/100*cval_stat[i]*cval_stat[j]
end
sm_stat = eigvecs(mcov_stat);
eival_stat = eigvals(mcov_stat)

#######
# systemalic
mcov_syst = one(cmsyst);
for i=1:size(cmsyst,1), j=1:size(cmsyst,2)
     mcov_syst[i,j] = cmsyst[i,j]/100*cval_syst[i]*cval_syst[j]
end
sm_syst = eigvecs(mcov_syst);
eival_syst = eigvals(mcov_syst);
```

```julia
# Pkg.add("Distributions")
using Distributions
#
function draw_stat()
    sm_stat*diagm([v>0 ? sqrt(v) : 0.0 for v in eival_stat])*rand(Normal(),Ncd)
end
function draw_syst()
    sm_syst*diagm([v>0 ? sqrt(v) : 0.0 for v in eival_syst])*rand(Normal(),Ncd)
end
```

```julia
diagm(eival_syst)
```

```julia
draw_stat()
```

```julia
begin
    newv = hcat([draw_stat() for i in 1:5000]...)
    mcor_stat_new = [cor(newv[i,:],newv[j,:]) for i=1:Ncd, j=1:Ncd]
    heatmap([v≈100.0 ? 0.0 : v for v in 100*mcor_stat_new])
end
```

```julia
begin
    newv = hcat([draw_syst() for i in 1:5000]...)
    mcor_new = [cor(newv[i,:],newv[j,:]) for i=1:Ncd, j=1:Ncd]
    heatmap([v≈100.0 ? 0.0 : v for v in 100*mcor_new])
end
```

```julia
plot(cval_s, cval_data+draw_stat()+draw_syst(),yerr=cval_stat,lw=0)
```
